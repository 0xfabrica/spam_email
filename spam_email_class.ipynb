# PreparaciÃ³n de Datos

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df) # Scale the data and assign to a new variable
df = pd.DataFrame(df_scaled, columns=df.columns) # Use the original DataFrame's columns
df.head()
```

```python
x = df.drop('class', axis=1)
y = df['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# CreaciÃ³n de Modelo RandomForestClassifier

```python
rf = RandomForestClassifier(n_estimators=50, max_depth=25)
rf.fit(x_train, y_train)
score = rf.score(x_train, y_train)
score # Overfitting
```

```python
y_pred = rf.predict(x_test)
```

## Matriz de confusiÃ³n

```python
cm  = confusion_matrix(y_test, rf.predict(x_test))
labels = np.array([
    [f"VN: {cm[0,0]}", f"FP: {cm[0,1]}"],
    [f"FN: {cm[1,0]}", f"VP: {cm[1,1]}"]
])

# Graficar la matriz de confusiÃ³n con seaborn
plt.figure(figsize=(6,4))
sns.heatmap(
            cm,
            annot=labels,  # anotaciones personalizadas
            fmt="",         # sin formato numÃ©rico adicional
            cmap="seismic")   # paleta de colores en azul

plt.title("Matriz de ConfusiÃ³n con Etiquetas TN/FP/FN/TP")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```

## Visualizar Primer Ãrbol

```python
from sklearn.tree import plot_tree

arbol_individual = rf.estimators_[0] # Cambia de numero para visualizar otros arboles del rf

# Visualiza el primer Ã¡rbol
plt.figure(figsize=(6, 4))
plot_tree(arbol_individual, filled=True, feature_names=x.columns)
plt.show()
```

## MÃ©tricas RandomForest

```python
print(classification_report(y_test, y_pred)) # MÃ©tricas del modelo en prueba
print(f"Score de entrenamiento:{score}") # MÃ©tricas del modelo en entrenamiento
# Overfitting
```

# Modelo XGBoost

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(n_estimators=100, random_state=42)
xgb_model.fit(x_train, y_train)

y_pred = xgb_model.predict(x_test)
```

## MÃ©tricas XGBoost

```python
# Calcular la precisiÃ³n
accuracy_xgb = accuracy_score(y_test, y_pred)
print("PrecisiÃ³n:", accuracy_xgb)

# Imprimir el informe de clasificaciÃ³n
print(classification_report(y_test, y_pred))
```

```python
cma  = confusion_matrix(y_test, xgb_model.predict(x_test))
labels = np.array([
    [f"VN: {cma[0,0]}", f"FP: {cma[0,1]}"],
    [f"FN: {cma[1,0]}", f"VP: {cma[1,1]}"]
])

# Graficar la matriz de confusiÃ³n con seaborn
plt.figure(figsize=(6,4))
sns.heatmap(
            cma,
            annot=labels,  # anotaciones personalizadas
            fmt="",         # sin formato numÃ©rico adicional
            cmap="seismic")   # paleta de colores en azul

plt.title("Matriz de ConfusiÃ³n con Etiquetas TN/FP/FN/TP")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```
